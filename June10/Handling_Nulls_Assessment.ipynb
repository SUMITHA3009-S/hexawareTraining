```python
bookings_df.na.fill({"booking_amount":0}).show()
```

```python
dq_df=bookings_df.withColumn("data_quality_status",
when(col("customer_name").isNull() | col("service_type").isNull() | col("booking_amount").isNull(),"Incomplete").otherwise("Complete"))
dq_df.show()
```

```python
dq_df.groupBy("data_quality_status").count().show()
```

```python
dq_df.filter(col("data_quality_status")=="Complete").show()
```

```python
dq_df.filter(col("data_quality_status")=="Incomplete").show()
```

```python
amount_df=bookings_df.na.fill({"booking_amount":0})
amount_df=amount_df.withColumn("tax",col("booking_amount")*0.05)
amount_df.show()
```

```python
amount_df=bookings_df.na.fill({"booking_amount":0})
amount_df=amount_df.withColumn("tax",col("booking_amount")*0.05).withColumn("final_amount",col("booking_amount")+col("tax"))
amount_df.show()
```

```python
amount_df=bookings_df.na.fill({"booking_amount":0})
amount_df.filter(col("booking_status")=="Confirmed").agg(sum("booking_amount").alias("revenue")).show()
```

```python
bookings_df.na.fill({"service_type":"Unknown"}).groupBy("service_type").count().show()
```

```python
bookings_df.na.fill({"city":"Unknown"}).groupBy("city").count().show()
```

```python
bookings_df.na.fill({"booking_amount":0}).agg(avg("booking_amount")).show()
```

```python
bookings_df.na.drop(subset=["booking_amount"]).agg(avg("booking_amount")).show()
```

```python
avg1=bookings_df.na.fill({"booking_amount":0}).agg(avg("booking_amount").alias("avg_with_zero"))
avg2=bookings_df.na.drop(subset=["booking_amount"]).agg(avg("booking_amount").alias("avg_without_null"))
avg1.show()
avg2.show()
```

```python
clean_bookings=bookings_df.na.fill({"city":"Unknown","provider":"Not Available","payment_mode":"Not Provided","booking_status":"Unknown","booking_amount":0})
clean_bookings.write.mode("overwrite").parquet("clean_bookings.parquet")
```
