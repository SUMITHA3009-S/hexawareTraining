```python
from pyspark.sql import SparkSession
spark = SparkSession.builder\
.appName("Working with files")\
.getOrCreate()
```

```python
df = spark.read.csv("flightBookings.csv",header=True,inferSchema=True)
df.show()
```

```python
df.show()
```

```python
df.select("passenger_name","from_city","to_city").show()
```

```python
from pyspark.sql.functions import *
df.filter(col("status")=="Confirmed").show()
```

```python
df.filter(col("status")=="Cancelled").show()
```

```python
df.filter(col("status")=="Pending").show()
```

```python
df.filter(col("ticket_price")>20000).show()
```

```python
df.filter(col("to_city").isin("Dubai","Singapore","London")).show()
```

```python
df.count()
```

```python
df.groupBy("airline").count().show()
```

```python
df.groupBy("status").count().show()
```

```python
df.filter(col("status")=="Confirmed").agg(sum("ticket_price").alias("revenue")).show()
```

```python
df.groupBy("airline").agg(avg("ticket_price").alias("avg_price")).show()
```

```python
df.agg(max("ticket_price").alias("highest_price")).show()
```

```python
df.agg(min("ticket_price").alias("lowest_price")).show()
```

```python
df.orderBy(col("ticket_price").desc()).show()
```

```python
df.withColumn("tax",col("ticket_price")*0.05).show()
```

```python
df.withColumn("tax",col("ticket_price")*0.05).withColumn("final_price",col("ticket_price")*1.05).show()
```

```python
df.withColumn("price_category",
when(col("ticket_price")>=30000,"Premium")
.when(col("ticket_price")>=10000,"Standard")
.otherwise("Budget")).show()
```

```python
df.withColumn("price_category",
when(col("ticket_price")>=30000,"Premium")
.when(col("ticket_price")>=10000,"Standard")
.otherwise("Budget")).groupBy("price_category").count().show()
```

```python
df.filter(col("status")=="Confirmed").write.mode("overwrite").parquet("confirmed_flights.parquet")
```

```python
# JSON ASSESSMENT

hotels_df = spark.read.option("multiline","true").json("hotels.json")
hotels_df.show(truncate=False)
hotels_df.printSchema()
```

```python
hotels_df.show(truncate=False)
```

```python
hotels_df.select("hotel_name","city","rating").show()
```

```python
hotels_df.filter(col("category")=="Luxury").show()
```

```python
hotels_df.filter(col("rating")>4.5).show()
```

```python
hotels_df.filter(col("rooms_available")>15).show()
```

```python
hotels_df.filter(col("price_per_night")>10000).show()
```

```python
hotels_df.filter(col("city").isin("Dubai","London")).show()
```

```python
hotels_df.filter(col("contact.phone").isNull()).show()
```

```python
hotels_df.filter(col("contact.email").isNull()).show()
```

```python
hotels_df.select("hotel_name","contact.phone","contact.email").show()
```

```python
hotels_df.filter(array_contains(col("amenities"),"wifi")).show()
```

```python
hotels_df.filter(array_contains(col("amenities"),"spa")).show()
```

```python
hotels_df.groupBy("city").count().show()
```

```python
hotels_df.groupBy("category").count().show()
```

```python
hotels_df.groupBy("category").agg(avg("rating").alias("avg_rating")).show()
```

```python
hotels_df.groupBy("city").agg(avg("price_per_night").alias("avg_price")).show()
```

```python
hotels_df.agg(max("price_per_night").alias("highest_price")).show()
```

```python
hotels_df.withColumn("total_potential_revenue",col("rooms_available")*col("price_per_night")).show()
```

```python
hotels_df.orderBy(col("rating").desc()).show()
```

```python
hotels_df.select(
"hotel_id","hotel_name","city","category","rating",
"rooms_available","price_per_night",
col("contact.phone").alias("phone"),
col("contact.email").alias("email")
).write.mode("overwrite").parquet("hotels_flattened.parquet")
```
