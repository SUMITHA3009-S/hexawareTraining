```python
from pyspark.sql import SparkSession
spark=SparkSession.builder\
.appName("RDD")\
.getOrCreate()
```

```python
numbers=[10,20,30,40,50]
rdd=spark.sparkContext.parallelize(numbers)
print(rdd.collect())
```

```python
rdd.count
```

```python
rdd.take(3)
```

```python
numbers=[10,20,30,40,50]
rdd=spark.sparkContext.parallelize(numbers)
result=rdd.map(lambda x:x*10)
print(result.collect())
```

```python
numbers=[1,2,3,4,5]
rdd=spark.sparkContext.parallelize(numbers)
result=rdd.map(lambda x:x*x)
print(result.collect())
```

```python
numbers=[1,2,3,4,5,6,7,8]
rdd=spark.sparkContext.parallelize(numbers)
result=rdd.filter(lambda x:x%2==0)
print(result.collect())
```

```python
numbers=[10,20,30,40,50]
rdd=spark.sparkContext.parallelize(numbers)
result=rdd.distinct()
print(result.collect())
```

```python
rdd1=spark.sparkContext.parallelize([10,20,30])
rdd2=spark.sparkContext.parallelize([40,50,60])
result=rdd1.union(rdd2)
print(result.collect())
```

```python
numbers=[40,10,60,20,50]
rdd=spark.sparkContext.parallelize(numbers)
result=rdd.sortBy(lambda x:x)
print(result.collect())
```

```python
numbers=[10,20,30,40]
rdd=spark.sparkContext.parallelize(numbers)
result=rdd.reduce(lambda x,y:x+y)
print(result)
```
