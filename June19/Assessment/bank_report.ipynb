```python
account_data = [
    (101, "Chennai", 50000),
    (102, "Chennai", 30000),
    (103, "Bangalore", 45000),
    (104, "Hyderabad", 70000),
    (105, "Bangalore", 25000)
]

columns = ["account_id", "city", "balance"]

bank_df = spark.createDataFrame(account_data, columns)

bank_df.createOrReplaceTempView("accounts")

report_df = spark.sql("""
SELECT city,
       SUM(balance) AS total_balance
FROM accounts
GROUP BY city
""")

display(report_df)

print("Bank Report Generated")
```
