```python
product_data = [
    (1, "Laptop", 50000),
    (2, "Mobile", 25000),
    (3, "Headphone", 3000),
    (4, "Monitor", 12000)
]

columns = ["product_id", "product_name", "price"]

products_df = spark.createDataFrame(product_data, columns)

display(products_df)

products_df.createOrReplaceTempView("products")

avg_price = spark.sql("""
SELECT AVG(price) AS average_price
FROM products
""")

display(avg_price)
```
