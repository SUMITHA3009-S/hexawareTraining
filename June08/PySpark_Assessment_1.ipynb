```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PySpark Day 1") \
    .getOrCreate()
```

```python
data = [
    (101, "Rahul Sharma", "Data Engineering", 75000, "Hyderabad"),
    (102, "Priya Reddy", "AI Engineering", 85000, "Bangalore"),
    (103, "Amit Kumar", "Data Engineering", 65000, "Mumbai"),
    (104, "Sneha Patel", "Data Science", 95000, "Chennai"),
    (105, "Farhan Ali", "Cloud Engineering", 80000, "Delhi")
]
```

```python
columns = [
    "employee_id",
    "name",
    "department",
    "salary",
    "city"
]
```

```python
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df = df.withColumn(
    "salary_category",
    when(df.salary >= 85000, "High")
    .when(df.salary >= 75000, "Medium")
    .otherwise("Low")
)
df.show()
```

```python
df.groupBy(df.salary_category).count().show()
```

```python
final_report = df.withColumn(
    "bonus",
    col("salary") * 0.10
).withColumn(
    "annual_salary",
    col("salary") * 12
)

final_report.select(
    "employee_id",
    "name",
    "department",
    "salary",
    "bonus",
    "annual_salary"
).show()
```
