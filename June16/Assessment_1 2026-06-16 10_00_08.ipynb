```python
source_df = df

etl_df = source_df.withColumn(
    "total_amount",
    col("quantity") * col("price")
)

completed_orders = etl_df.filter(
    col("status") == "Completed"
)

completed_orders.show()
```

```python
city_revenue_report = completed_orders.groupBy(
    "city"
).agg(
    sum("total_amount").alias("city_revenue")
)

city_revenue_report.show()
```

```python
category_revenue_report = completed_orders.groupBy(
    "category"
).agg(
    sum("total_amount").alias("category_revenue")
)

category_revenue_report.show()
```

```python
completed_orders.write.mode(
    "overwrite"
).parquet("/dbfs/sales_parquet")

parquet_df = spark.read.parquet('/dbfs/sales_parquet')

parquet_df.show()
```

```python
parquet_df.createOrReplaceTempView(
    "sales_parquet_view"
)
```

```python
spark.sql("""
select 
category,
count(*) as total_orders,
sum(total_amount) as total_revenue
from sales_parquet_view
group by category
order by total_revenue desc
""").show()
```
