```python
df_csv_patients = spark.read.csv("patients.csv", header=True, inferSchema=True)
df_csv_appointments = spark.read.csv("appointments.csv", header=True, inferSchema=True)
```

```python
df_json_preferences = spark.read.option("multiline", "true").json("patient_preferences.json")
```

```python
df_clean_patients = df_csv_patients.na.fill({"city": "Unknown", "blood_group": "Not Available"})
df_clean_appointments = df_csv_appointments.na.fill({"consultation_fee": 0})
```

```python
from pyspark.sql.functions import col
df_intermediate = df_clean_patients.join(df_clean_appointments, "patient_id", "left")
preferences_flat = df_json_preferences.select(
    col("patient_id").alias("json_join_id"),
    col("contact.email").alias("contact_email"),
    col("contact.phone").alias("contact_phone"),
    col("preferred_hospital")
)

df_joined = df_intermediate.join(
    preferences_flat,
    df_intermediate["patient_id"] == preferences_flat["json_join_id"],
    "left"
).drop("json_join_id").cache()
```

```python
df_etl = df_joined.withColumn("age_group", when(col("age") < 30, "Young").when((col("age") >= 30) & (col("age") <= 50), "Middle").otherwise("Senior"))
```

```python
df_metrics = df_etl.withColumn("revenue_contribution",when(col("status") == "Completed",col("consultation_fee")).otherwise(0))
```

```python
df_patient_spending = df_metrics.groupBy(df_metrics['patient_id'], "patient_name").agg(sum("revenue_contribution").alias("total_spent"))
df_patient_spending.show()
```

```python
df_dept_revenue = df_metrics.filter(col("department").isNotNull()).groupBy("department").agg(sum("revenue_contribution").alias("total_revenue"))
df_dept_revenue.show()
```

```python
df_metrics.write.mode("overwrite").parquet("final_hospital_analytics.parquet")
```

```python
print("=== Hospital Analytics Report ===")
print(f"Total Patients Analyzed: {df_clean_patients.count()}")
print(f"Total Appointments Processed: {df_clean_appointments.count()}")
print("Top Revenue Departments:")
df_dept_revenue.orderBy(desc("total_revenue")).show(5)
```
